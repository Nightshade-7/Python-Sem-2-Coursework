```python
#Problem 1

#Part 1

def gcd(a, b): # We assyme that the inputs are always positive, aka a,b > 0
    while a != b: 
        if a > b:
            a = a - b
        else:
            b = b - a
    return a
```

```python
#Problem 1

print(gcd(500,1000))

#Part 2

print()
print("If one or both arguments are negative, the function displays an error.")
```

```python
#Problem 2

#Part 1

def euler_limit(n):
    a=(1 + 1/n)**n
    return a

print(euler_limit(6789))
```

```python
#Problem 2

#Part 3

def factorial(n):
    final=1
    if n==0:
        return final
    for i in range(1,n+1):
        final*=i
    return final

def euler_sum(n):
    total=0
    for i in range(0,n+1,1):
        total += 1/factorial(i)
    return total

print(euler_sum(50))
```

```python
#Problem 2

#Part 4

def euler_approx(n:float):
    import math
    final=0
    m=0
    while math.exp(1)-euler_sum(m)> n:
        m= m+1
        final= euler_sum(m)
    return final

print(euler_approx(0.5))
```

```python
#Problem 3

#Part 1

def is_power_of_2(n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if 2**i==n:
                return True
        return False

is_power_of_2(65)
```

```python
#Problem 3

#Part 2

def is_power(b,n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if b**i==n:
                return True
        return False

is_power(10,64)
```
